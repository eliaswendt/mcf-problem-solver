```python
import pandas as pd
```

```python
df = pd.read_csv("simulated_annealing.csv")
print("Final cost:", list(df['cost'])[-1])
print("Final delay cost:", list(df['delay_cost'])[-1])
print("Final edge cost:", list(df['edge_cost'])[-1])
print("Final travel cost:", list(df['travel_cost'])[-1])
df.plot(y=['cost', 'delay_cost', 'edge_cost', 'travel_cost'], x='time')
df.plot(y=['travel_cost'], x='time', logy=True)
df.plot(y=['temperature'], x='time', logy=True)
```
